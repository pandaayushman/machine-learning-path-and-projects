```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/parkinsons_kaggle.csv")

print(df.head())
print(df.describe())
print(df.isnull().sum())

sns.pairplot(df)
plt.show()

sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.show()
```
